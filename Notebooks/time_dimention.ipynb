```python
# Import utility libraries.
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Run if you use dark theme.
mpl.rcParams['text.color'] = 'black'
mpl.rcParams['axes.labelcolor'] = 'black'
mpl.rcParams['xtick.color'] = 'black'
mpl.rcParams['ytick.color'] = 'black'
```

```python
# Seaborn font size.
sns.set(font_scale=1)
```

```python
# Visualizer for all included searchs by size.
def print_searchs_comparison_size(test):
    for search_type in set(test["search_type"]):
        search_df = test[test['search_type'] == search_type]
        graph_types = search_df['graph_type'].unique()

        plt.figure(figsize=(20, 10))

        for graph_type in graph_types:
            current_data = search_df[search_df['graph_type'] == graph_type]
            plt.plot(current_data['size'], current_data['time(ns)'], label=graph_type)

        plt.title(search_type)
        plt.xlabel('Graph Size')
        plt.xticks(search_df['size'].unique())
        plt.ylabel('Time(ns)')
        plt.legend(labelcolor='black', prop={'size': 15})
```

```python
# Visualizer for all included searchs by edges.
def print_searchs_comparison_edges(test):
    for search_type in set(test["search_type"]):
        search_df = test[test['search_type'] == search_type]
        graph_types = search_df['graph_type'].unique()

        plt.figure(figsize=(20, 10))

        for graph_type in graph_types:
            current_data = search_df[search_df['graph_type'] == graph_type]
            plt.plot(current_data['edges'], current_data['time(ns)'], label=graph_type)

        plt.title(search_type)
        plt.xlabel('Graph Edges')
        plt.xticks(search_df['edges'].unique())
        plt.ylabel('Time(ns)')
        plt.legend(labelcolor='black', prop={'size': 15})
```

```python
# Visualizer for all 4 types of generated arrays by size.
def print_graphs_comparison_size(test):
    graph_types = test['graph_type'].unique()

    fig, axes = plt.subplots(2, 2, figsize=(25, 25))

    flattened_axes = axes.flatten()

    for i in range(graph_types.shape[0]):
        ax = flattened_axes[i]
        graph_type = graph_types[i]

        current_df = test[test['graph_type'] == graph_type]

        search_types = current_df['search_type'].unique()
        for search_type in search_types:
            current_df_grouped_by_search_type = current_df[current_df['search_type'] == search_type]
            ax.plot(current_df_grouped_by_search_type['size'], current_df_grouped_by_search_type['time(ns)'], label=search_type)

        ax.set_title(graph_type)
        ax.set_xlabel('Graph Size')
        ax.set_ylabel('Time(ns)')
        ax.legend()
        ax.plot()
```

```python
# Visualizer for all 4 types of generated arrays by edges.
def print_graphs_comparison_edges(test):
    graph_types = test['graph_type'].unique()

    fig, axes = plt.subplots(2, 2, figsize=(25, 25))

    flattened_axes = axes.flatten()

    for i in range(graph_types.shape[0]):
        ax = flattened_axes[i]
        graph_type = graph_types[i]

        current_df = test[test['graph_type'] == graph_type]

        search_types = current_df['search_type'].unique()
        for search_type in search_types:
            current_df_grouped_by_search_type = current_df[current_df['search_type'] == search_type]
            ax.plot(current_df_grouped_by_search_type['edges'], current_df_grouped_by_search_type['time(ns)'], label=search_type)

        ax.set_title(graph_type)
        ax.set_xlabel('Graph Edges')
        ax.set_ylabel('Time(ns)')
        ax.legend()
        ax.plot()
```

```python
# Data reading from 'test_time.csv' in 'Tests' folder.
test = pd.read_csv('../Tests/test_time.csv', sep=';', header=None)
test.columns = ['graph_type', 'size', 'edges', 'search_type', 'time(ns)']
```

---

```python
print_searchs_comparison_size(test)
```

```python
print_searchs_comparison_edges(test)
```

---

```python
print_graphs_comparison_size(test)
```

```python
print_graphs_comparison_edges(test)
```
